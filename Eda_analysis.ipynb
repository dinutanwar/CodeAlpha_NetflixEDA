# 📊 Netflix Movies & TV Shows - Exploratory Data Analysis (EDA)

# Import Libraries

```python

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
df = pd.read_csv("C:\\Users\\DINESH\\OneDrive\\Documents\\Netflix.csv")
```

```python
df.head()
```

# Dataset Information

```python
# Dataset Information
print("\nDataset Info:")
print(df.info())
```

# Check Missing Values

```python
# Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

# -------------------------------
# Data Cleaning
# -------------------------------

```python
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Not Rated")
df['director'] = df['director'].fillna("No Director")
```

# -------------------------------
# Basic Analysis
# -------------------------------

```python
# Total Movies and TV Shows
print("\nContent Type Count:")
print(df['type'].value_counts())
```

# Top 10 Countries



```python
# Top 10 Countries
top_countries = df['country'].value_counts().head(10)
```

```python
# Top 10 Ratings
top_ratings = df['rating'].value_counts().head(10)
```

```python

# Release Year Trend
release_year = df['release_year'].value_counts().sort_index()
```

```python
# -------------------------------
# Visualization 1
# Movies vs TV Shows
# -------------------------------
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

```python
# -------------------------------
# Visualization 2
# Top 10 Countries
# -------------------------------
```

```python
plt.figure(figsize=(10,6))
top_countries.plot(kind='bar')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
# -------------------------------
# Visualization 3
# Ratings Distribution
# -------------------------------
```

```python
plt.figure(figsize=(10,6))
top_ratings.plot(kind='bar')
plt.title("Top Netflix Ratings")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
release_year_counts = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(14,6))
plt.bar(release_year_counts.index, release_year_counts.values, color='skyblue', edgecolor='black')
plt.title("Netflix Content Release Trend", fontsize=16)
plt.xlabel("Release Year", fontsize=12)
plt.ylabel("Number of Releases", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
